# Visualisation
## Univariate analysis

```python
import seaborn as sns
```

```python
sns.displot(data['mpg'],bins=6)
```

```python
sns.boxplot(y='mpg',data=data)
#we can see that it has an outlier
```

```python
#count the frequency of categories
sns.countplot(x='gear',data=data)
```

```python
#shows the density of data(Similar to kde plot)-just another way of representing data
sns.violinplot(y='mpg',data=data)
#A violin plot is a data visualization tool that combines the features of a box plot and a kernel density plot to show the distribution of a dataset.
#The plot is usually mirrored on both sides of a central vertical axis, making it easier to compare distributions visually.
```

## Bivariate analysis

```python
#numeric-numeric->scatterplot

sns.lmplot(x='mpg',y='hp',data=data,hue='am')
```

```python
#categorical-categorical
sns.countplot(x='cyl',data=data,hue='am')
```

```python
#categorical-numerical->boxplot
sns.boxplot(x='cyl',y='mpg',data=data)
#as we can see the third one has outliers.the first one has median like 26 q3 is 30 and q1 is 23
```
